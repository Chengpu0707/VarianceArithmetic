```python
fig, ax = plt.subplots()
ax.set_title('Matrix Determinant Precision vs Condition Number')
ax.set_xlabel('Condition Number')
ax.set_xscale('log')
ax.set_ylabel('Determinant Precision')
ax.set_yscale('log')
plt.grid(True)

for size in range(4, 10):
    dfPlot = df[(df['Type']=='Random') & (df['Size']==size) & (df['Noise']==1e-15)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size}')
#ax.plot(dfMinus['Order'], dfMinus['Uncertainty'], color+'--', label=f'+ {1 - epsilon} Uncertainty')

ax.legend(bbox_to_anchor=(1.0, 1.0))
```

```python
fig, ax = plt.subplots()
ax.set_title('Matrix Determinant Precision vs Condition Number')
ax.set_xlabel('Condition Number')
ax.set_xscale('log')
ax.set_ylabel('Determinant Precision')
ax.set_yscale('log')
plt.grid(True)

for size, color, line in ((6,'k','--'), (7,'k',':'), (8,'k','-')):
    dfPlot = df[(df['Type']=='Random') & (df['Size']==size) & (df['Noise']==1e-16)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size} Random 1e-16',
            color=color, linestyle=line)
for size, color, line in ((6,'gray','--'), (7,'gray',':'), (8,'gray','-')):
    dfPlot = df[(df['Type']=='Random') & (df['Size']==size) & (df['Noise']==1e-15)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size} Random 1e-15',
            color=color, linestyle=line)
for size, color, marker in ((4,'k','x'), (5,'k','+'), (6,'k','.'), (7,'k','*')):
    dfPlot = df[(df['Type']=='Hilbert') & (df['Size']==size) & (df['Noise']==0)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size} Random 0',
            color=color, marker=marker, linestyle='')
for size, color, marker in ((4,'gray','x'), (5,'gray','+'), (6,'gray','.'), (7,'gray','*')):
    dfPlot = df[(df['Type']=='Hilbert') & (df['Size']==size) & (df['Noise']==1e-15)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size} Random 1e-15',
            color=color, marker=marker, linestyle='')
ax.legend(bbox_to_anchor=(1.0, 1.0))
```
