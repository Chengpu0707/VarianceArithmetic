```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats as spst
```

```python
import os
print(os.getcwd())
```

```python
def Plot6Density(sX:np.array, ssY:list[np.array], 
                 title:str, xlabel:str, 
                 sColor=['gray','black']*3, sLineStyle=['--',':','-']*2,
                 ylim=None):
    sSum = [np.sum(sY) for sY in ssY]
    for idx, sY, c, ls  in zip(range(6), ssY, sColor, sLineStyle):
        plt.plot(sX, sY/sSum[idx]/(sX[-1]-sX[-2]), label=f'$\mu$={idx}', c=c, ls=ls)
    plt.legend()
    plt.grid()
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Probability Density')
    if ylim:
        plt.ylim(ylim)
```

# Exponent

```python
sX = np.linspace(-4,8,1200)
ssY = [spst.norm.pdf(sX, mu, 1)/np.exp(sX) for mu in (0, 1, 2, 3, 4, 5)]
```

```python
Plot6Density(sX, ssY, r'$\tilde{y} = e^{\tilde{x} = \mu \pm 1}$', r'$\log(\tilde{y})$')
```

# Log

```python
sX = np.linspace(0,9,900)
ssY = [spst.norm.pdf(sX, mu, 1)*sX for mu in (0, 1, 2, 3, 4, 5)]
```

```python
Plot6Density(sX, ssY, r'$\tilde{y} = log(\tilde{x} = \mu \pm 1)$', r'$e^\tilde{y}$')
```

# Power

```python
sX = np.linspace(0,9,9000)
ssY = [spst.norm.pdf(sX, mu, 1)*np.sqrt(sX) for mu in (0, 1, 2, 3, 4, 5)]
```

```python
Plot6Density(sX, ssY, r'$\tilde{y} = \sqrt{\tilde{x} = \mu \pm 1}$', r'$\tilde{y}^2$')
```

```python
sX = np.linspace(0.01,9,8990)
ssY = [spst.norm.pdf(sX, mu, 1)/sX for mu in (0, 1, 2, 3, 4, 5)]
```

```python
Plot6Density(sX, ssY, r'$\tilde{y} = (\tilde{x} = \mu \pm 1)^2$', r'$\sqrt{\tilde{y}}$', ylim=(0,0.6))
```

```python
sX = np.linspace(-5,9,1500)
ssY = [spst.norm.pdf(sX, mu, 1)*sX*sX for mu in (0, 1, 2, 3, 4, 5)]
```

```python
Plot6Density(sX, ssY, r'$\tilde{y} = 1/(\tilde{x} = \mu \pm 1)$', r'$1/\tilde{y}$')
```
