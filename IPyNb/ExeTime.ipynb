```python
dfJ = pd.read_csv("../Java/Output/FFT_2_19.txt.log.txt", sep='\t')
dfJ.head()
```

```python
ax = dfP.plot(x='Order', y='Mean', label='Python', style='k:')
dfC.plot(ax=ax, x='Order', y='Mean', label='C++ -O3', style='k--')
dfJ.plot(ax=ax, x='Order', y='Mean', label='Java', style='k-', grid=True, 
         ylabel='Execution time [seconds]', logy=True, title='Performance Comparison Using FFT')
```
