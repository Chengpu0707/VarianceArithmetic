```python
df = pd.read_csv("../Python/Output/ExpLogError.txt", sep='\t')
df.head()
```

```python
df['Normalized Error'] = df['Error'] / df['Uncertainty']
df.head()
```

```python
df[df['Type'] == 'exp(log(x))'].head()
```

```python
ax = df[df['Type'] == 'log(exp(x))'].plot(x='X', y='Error', color='gray', style='-', label='log(exp(x)) value error')
df[df['Type'] == 'exp(log(x))'].plot(x='X', y='Error', color='black', style='-', label='exp(log(x)) value error', ax=ax)
df[df['Type'] == 'log(exp(x))'].plot(x='X', y='Uncertainty', color='gray', style='--', label='log(exp(x)) uncertainty', ax=ax)
df[df['Type'] == 'exp(log(x))'].plot(x='X', y='Uncertainty', color='black', style='--', label='exp(log(x)) uncertainty', ax=ax, grid=True)
```
