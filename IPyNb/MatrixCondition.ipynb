```python
fig, ax = plt.subplots()
ax.set_title('Matrix Determinant Precision vs Condition Number')
ax.set_xlabel('Condition Number')
ax.set_xscale('log')
ax.set_ylabel('Determinant Precision')
ax.set_yscale('log')
plt.grid(True)

for size in range(4, 9):
    dfPlot = df[(df['Type']=='Random') & (df['Size']==size)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size}')
#ax.plot(dfMinus['Order'], dfMinus['Uncertainty'], color+'--', label=f'+ {1 - epsilon} Uncertainty')

ax.legend(bbox_to_anchor=(1.0, 1.0))
```

```python
fig, ax = plt.subplots()
ax.set_title('Matrix Determinant Precision vs Condition Number')
ax.set_xlabel('Condition Number')
ax.set_xscale('log')
ax.set_ylabel('Determinant Precision')
ax.set_yscale('log')
plt.grid(True)

for size, color, line in ((4,'gray','--'), (5,'gray','-'), (6,'k','--'), (7,'k',':'), (8,'k','-')):
    dfPlot = df[(df['Type']=='Random') & (df['Size']==size)].sort_values('Condition Number')
    ax.plot(dfPlot['Condition Number'], dfPlot['Determinant Precision'], label=f'{size} Random',
            color=color, linestyle=line)
dfPlot = df[(df['Type']=='Hilbert') & (df['Size']==3) & (df['Noise']==0)].sort_values('Condition Number')
ax.scatter(dfPlot['Condition Number'], dfPlot['Determinant Precision'], color='k', marker='.', label='3 Hilbert')
dfPlot = df[(df['Type']=='Hilbert') & (df['Size']==4) & (df['Noise']==0)].sort_values('Condition Number')
ax.scatter(dfPlot['Condition Number'], dfPlot['Determinant Precision'], color='k', marker='x', label='4 Hilbert')
dfPlot = df[(df['Type']=='Hilbert') & (df['Size']==5) & (df['Noise']==0)].sort_values('Condition Number')
ax.scatter(dfPlot['Condition Number'], dfPlot['Determinant Precision'], color='gray', marker='+', label='5 Hilbert')

ax.legend(bbox_to_anchor=(1.0, 1.0))
```
