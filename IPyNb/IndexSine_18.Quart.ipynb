```python
sHisto = [str(v/10) for v in range(-20, 21, 2)]
print(sHisto)
```

```python
dfC[sHisto].head()
```

```python
dfC.set_index(dfC['Context'])
dfJ[sHisto].head()
```

```python
ax = dfC[sHisto].set_index(dfC['Context']).T.plot(c='gray', ls='-')
dfJ[sHisto].set_index(dfJ['Context']).T.plot(ax=ax, c='black', ls=':')
dfCJ[sHisto].set_index(dfCJ['Context']).T.plot(ax=ax, style='kx', grid=True, 
                                               xlabel='Difference of LSB', ylabel='Normalized Count')
```

```python
ax = dfC[sHisto].set_index(dfC['Context']).T.plot(style='kx')
dfJ[sHisto].set_index(dfJ['Context']).T.plot(ax=ax, style='k+')
dfCJ[sHisto].set_index(dfCJ['Context']).T.plot(ax=ax, style='k.', grid=True, logy=True, 
                                               xlabel='Difference of LSB', ylabel='Normalized Count')
```
