```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def corr(prec:float, corr_unc:float):
    return 1/((1/corr_unc - 1)* prec**2 + 1)
```

```python
x = np.arange(0,1, 0.01)
```

```python
plt.plot(x, corr(x, 0.01), label='0.01', c='k', ls='-')
plt.plot(x, corr(x, 0.02), label='0.02', c='k', ls='--')
plt.plot(x, corr(x, 0.05), label='0.05', c='k', ls=':')
plt.plot(x, corr(x, 0.10), label='0.10', c='gray', ls='-')
plt.title('Correlations of two inputs vs input uncertainties')
plt.xlabel('Precision')
plt.xscale('log')
plt.ylabel('Correlation of two inputs')
plt.yscale('log')
plt.legend()
plt.grid()
```
