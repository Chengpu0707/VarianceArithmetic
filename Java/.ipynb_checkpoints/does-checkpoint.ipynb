```python
def PlotValues( sValue=['Dev', 'Uncertainty'], sStyle=['x', '-'],
                sNoise=sorted(df['Noise'].unique()), noiseType='Uniform', 
                logx=True, logy=True, xLim=[], yLim=[] ):
    fig, ax = plt.subplots()
    if logx:
        ax.set_xscale('log')
    if logy:
        ax.set_yscale('log')
    if xLim:
        ax.set_xlim(yLim)
    if yLim:
        ax.set_ylim(yLim)
    for noise in sNoise:
        for value, style in zip(sValue, sStyle):
            dfPlot = df[(df['NoiseType']==noiseType) & (df['Noise']==noise)]
            ax.plot(dfPlot['X'], dfPlot[value], style, label=f'{noiseType} {noise} {value}')
    ax.legend(bbox_to_anchor=(1.1, 1.05))
```

# Uncertainty vs Dev

```python
PlotValues()
```

```python
PlotValues(yLim=[math.pow(10,-3), math.pow(10,1)])
```

```python
PlotValues(yLim=[math.pow(10,-3), math.pow(10,1)], noiseType='Gaussian')
```

# Mean vs Bias vs Mode

```python
PlotValues(sValue=['Mean', 'Bias', 'Mode'], sStyle=['-', '--', 'o'], logy=False, yLim=[-0.1, 0.1])
```
